```python
class person:
  def __init__(self,name,age):
    self.name = name
    self.age = age

p1 = person("Johm","32")

print(p1.name, p1.age)
```

```python
class dog:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def bark(self):
    print(self.name + " says woof")

d1 = dog("buddy", 3)
#changing property
d1.age = 5
#deleting the property
del d1.age

print(d1.age)


d1.bark()
```

```python
class car:
  def __init__(self, brand):
    self.brand = brand
  def show(self):
    print(self.brand)
c1 = car("Ford")

c1.show()
```

```python
class student:
  def __init__(self, name, grade):
    self.name = name
    self.grade = grade
p1 = student("Anna", "A")

print(p1.grade)

p1.grade = "B"

print(p1.grade , ' = After change')
```

```python
class Rectangle:
  def area(self,l,b):
    return l*b

  def perimeter(self,l,b):
    a = l+b
    return 2*a

calc = Rectangle()
print(calc.area(5,6))
print(calc.perimeter(5,6))

```
