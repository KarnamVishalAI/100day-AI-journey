```python
#Averag of three numbers

def avg_num(a,b,c):
  d = a+b+c
  avg = d/3
  return avg

avg_num(10,20,30)
```

```python
#odd or even
def odd_eve(a):
  if a%2 == 0:
    print("even")
  else:
    print("odd")

odd_eve(15)
```

```python
def find_max(numbers):
    return sorted(numbers)[-1]
print(find_max([3, 8, 1, 9, 4]))
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
print(count_vowels("Hello"))
```
