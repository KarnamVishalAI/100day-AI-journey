```python
import os
os.mkdir("resumes")
```

```python
open("resumes/resume1.txt","w").write("I am good at python and data science. I know numpy and pandas.")
open("resumes/resume2.txt","w").write("Machine learning enthusiast. Worked on tensorflow deep learning projects.")
open("resumes/resume3.txt","w").write("I have experience in web development and html css javascript.")
```

```python
open("job_description.txt","w").write("We are looking for a machine learning intern with python, numpy, pandas, tensorflow and deep learning skills.")
```

```python
import os
import string
import datetime
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def clean(t):
    t = t.lower()
    t = "".join(i for i in t if i not in string.punctuation)
    return t

skills = ["python","machine learning","deep learning","data science","tensorflow","pandas","numpy","sql","ai","nlp"]

resumes = []
names = []
resume_skills = []

for file in os.listdir("resumes"):
    with open("resumes/" + file, encoding="utf-8") as f:
        text = clean(f.read())
        resumes.append(text)
        names.append(file)
        resume_skills.append([s for s in skills if s in text])

with open("job_description.txt", encoding="utf-8") as f:
    job = clean(f.read())

docs = [job] + resumes
vec = TfidfVectorizer(stop_words="english")
mat = vec.fit_transform(docs)

job_vec = mat[0]
res_vecs = mat[1:]

sim = cosine_similarity(job_vec, res_vecs)[0]

final = []

for i in range(len(names)):
    skill_score = len(resume_skills[i]) / len(skills)
    total = sim[i]*0.7 + skill_score*0.3
    final.append((names[i], total, resume_skills[i]))

final.sort(key=lambda x: x[1], reverse=True)

print("\nAI Resume Screening System\n")

for n, s, sk in final:
    print(n, "-", round(s*100,2), "% Match")
    print("Matched Skills:", sk, "\n")

report = "AI Resume Screening Report\nDate: " + str(datetime.date.today()) + "\n\n"

for n, s, sk in final:
    report += n + " - " + str(round(s*100,2)) + "% Match\nSkills: " + ", ".join(sk) + "\n\n"

open("report.txt","w").write(report)

print("Report Generated: report.txt")
print("Final Selection:", final[0][0])
```
