```python
import random

print("WELCOME TO MINI GAME HUB")
name = input("Enter your name: ")

def rock_paper_scissors():
    while True:
        print("\nROCK PAPER SCISSORS")
        user = input("rock / paper / scissors or back: ").lower()

        if user == "back":
            break

        if user != "rock" and user != "paper" and user != "scissors":
            print("Wrong choice")
            continue

        computer = random.choice(["rock","paper","scissors"])
        print("Computer:", computer)

        if user == computer:
            print("Tie")
        elif user == "rock" and computer == "scissors":
            print("You Win")
        elif user == "paper" and computer == "rock":
            print("You Win")
        elif user == "scissors" and computer == "paper":
            print("You Win")
        else:
            print("Computer Wins")

def guess_the_number():
    number = random.randint(1,20)
    count = 0

    while True:
        user = input("Guess number 1-20 or back: ")

        if user == "back":
            break

        if not user.isdigit():
            print("Enter number")
            continue

        user = int(user)
        count = count + 1

        if user == number:
            print("Correct in",count,"tries")
            break
        elif user > number:
            print("Too High")
        else:
            print("Too Low")

def dice_battle():
    print("\n1 Single Player")
    print("2 Multiplayer")
    mode = input("Choose: ")

    if mode == "1":
        you = random.randint(1,6)
        system = random.randint(1,6)

        print("You rolled:", you)
        print("System rolled:", system)

        if you > system:
            print("You Win")
        elif you < system:
            print("System Wins")
        else:
            print("Tie")

    if mode == "2":
        p1 = input("Player 1 name: ")
        p2 = input("Player 2 name: ")

        r1 = random.randint(1,6)
        r2 = random.randint(1,6)

        print(p1,"rolled:",r1)
        print(p2,"rolled:",r2)

        if r1 > r2:
            print(p1,"Wins")
        elif r2 > r1:
            print(p2,"Wins")
        else:
            print("Tie")

while True:
    print("\nMAIN MENU")
    print("1 Rock Paper Scissors")
    print("2 Guess The Number")
    print("3 Dice Battle")
    print("4 Exit")

    choice = input("Choose: ")

    if choice == "1":
        rock_paper_scissors()
    elif choice == "2":
        guess_the_number()
    elif choice == "3":
        dice_battle()
    elif choice == "4":
        print("Thanks for playing",name)
        break
    else:
        print("Wrong choice")
```
